```python
# Import libraries
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

# Import dataset
# Sesuaikan dengan lokasi file xclara.csv di local komputer
# URL dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/lenses/lenses.data"
names = ['Age', 'Spectacle Prescription', 'Astigmatism', 'Tear Production Rate', 'Class']
data = pd.read_csv(url, names=names, delim_whitespace=True)

print(data.shape)
print(data.head(20))
```

```python
# Plot dataset
f1 = data['Age'].values
f2 = data['Spectacle Prescription'].values
X = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=7,)
#Kode memplot data
```

```python
# Plot dataset
f1 = data['Age'].values
f2 = data['Spectacle Prescription'].values
f5 = data['Class'].values
X = np.array(list(zip(f1, f2, f5)))
plt.scatter(f1, f2, c=f5, s=7)
#Kode memplot data
```

```python
# KMeans clustering
from sklearn.cluster import KMeans

# Menentukan jumlah cluster
kmeans = KMeans(n_clusters=4)

# Fitting input data
kmeans = kmeans.fit(X)

# Mendapatkan cluster labels
labels = kmeans.predict(X)

# Mendapatkan nilai centroid
C = kmeans.cluster_centers_

# Mencetak nilai centroid
print(C)
#Kode melakukan clustering dengan KMeans untuk membagi data menjadi 4 cluster.
```

```python
from sklearn.cluster import KMeans

for k in range(1, 10):
    # Menentukan jumlah cluster
    kmeans = KMeans(n_clusters=k, random_state=1)

    # Fitting input data
    kmeans = kmeans.fit(X)

    # Mendapatkan cluster labels
    labels = kmeans.predict(X)

    # Menghitung jumlah jarak antara setiap sampel dengan cluster centroid-nya (SSE)
    inertia = kmeans.inertia_

    print(f"k: {k}, cost: {inertia}")

print()
#Kode ini melakukan evaluasi untuk menentukan jumlah cluster optimal dengan KMeans.
```

```python
from sklearn.metrics.cluster import silhouette_score
silhouette_score(X, labels)
#Kode ini menghitung nilai silhouette score untuk mengevaluasi kualitas clustering.
```
