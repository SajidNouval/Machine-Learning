```python
#Menggunakan Decision Tree
import pandas
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# URL dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/lenses/lenses.data"
names = ['Age', 'Spectacle Prescription', 'Astigmatism', 'Tear Production Rate', 'Class']
dataset = pandas.read_csv(url, names=names, delim_whitespace=True)

# Memisahkan fitur (X) dan label (Y)
array = dataset.values
X = array[:, 0:4]
Y = array[:, 4]

# Membagi dataset menjadi data latih dan validasi
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(
    X, Y, test_size=validation_size, random_state=seed
)

# Membuat model Decision Tree
dt = DecisionTreeClassifier(random_state=seed)
dt.fit(X_train, Y_train)  # Melatih model

# Membuat prediksi pada data validasi
predictions_dt = dt.predict(X_validation)

# Evaluasi hasil
print("Evaluasi Model Decision Tree:")
print("=" * 40)
print(f"Accuracy Score    : {accuracy_score(Y_validation, predictions_dt):.3f}")
print("Confusion Matrix :")
print(confusion_matrix(Y_validation, predictions_dt))
print("\nClassification Report:")
print(classification_report(Y_validation, predictions_dt, zero_division=1))
```

```python
#Menggunakan Random Forest
import pandas
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# URL dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/lenses/lenses.data"
names = ['Age', 'Spectacle Prescription', 'Astigmatism', 'Tear Production Rate', 'Class']
dataset = pandas.read_csv(url, names=names, delim_whitespace=True)

# Memisahkan fitur (X) dan label (Y)
array = dataset.values
X = array[:, 0:4]
Y = array[:, 4]

# Membagi dataset menjadi data latih dan validasi
validation_size = 0.20
seed = 7
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(
    X, Y, test_size=validation_size, random_state=seed
)

# Membuat model Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=seed)
rf.fit(X_train, Y_train)  # Melatih model

# Membuat prediksi pada data validasi
predictions_rf = rf.predict(X_validation)

# Evaluasi hasil
print("Evaluasi Model Random Forest:")
print("=" * 40)
print(f"Accuracy Score    : {accuracy_score(Y_validation, predictions_rf):.3f}")
print("Confusion Matrix :")
print(confusion_matrix(Y_validation, predictions_rf))
print("\nClassification Report:")
print(classification_report(Y_validation, predictions_rf, zero_division=1))
```

Cari dataset lain dari repository UCI Machine Learning
- lakukan keenam langkah untuk menampilkan summary dataset pada modul 1
- ⁠kemudian buat evaluasi algoritma dengan menggunakan 3 model yaitu KNN, NB, dan SVM


```python
import pandas
from sklearn import model_selection
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# URL dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/lenses/lenses.data"
# Nama kolom dataset
names = ['Age', 'Spectacle Prescription', 'Astigmatism', 'Tear Production Rate', 'Class']
# Membaca dataset dari URL menggunakan pandas
# `delim_whitespace=True` digunakan karena data dipisahkan oleh spasi
dataset = pandas.read_csv(url, names=names, delim_whitespace=True)

# Memisahkan fitur (X) dan label (Y)
# array[:, 0:4] -> mengambil kolom fitur
# array[:, 4]   -> mengambil kolom label (kelas)
array = dataset.values
X = array[:, 0:4]  # Fitur (kolom 0-3)
Y = array[:, 4]    # Label (kolom 4)

# Membagi dataset menjadi data latih (80%) dan data validasi (20%)
validation_size = 0.20  # 20% data untuk validasi
seed = 7  # Seed untuk memastikan hasil pembagian konsisten
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(
    X, Y, test_size=validation_size, random_state=seed
)

# Daftar model untuk diuji
models = [
    ('KNN', KNeighborsClassifier()),  # K-Nearest Neighbors
    ('Naive Bayes', GaussianNB()),    # Naive Bayes
    ('SVM', SVC())                    # Support Vector Machine
]

# Evaluasi setiap model menggunakan cross-validation
print("Evaluasi Model dengan Cross-Validation:")
print("=" * 40)
scoring = 'accuracy'  # Metode evaluasi: akurasi
for name, model in models:
    # Menggunakan 10-fold cross-validation untuk mengevaluasi akurasi model
    kfold = model_selection.KFold(n_splits=10, random_state=seed, shuffle=True)
    cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
    # Menampilkan hasil evaluasi dengan rata-rata akurasi dan standar deviasi
    print(f"{name:12}: Mean Accuracy = {cv_results.mean():.3f} (Std = {cv_results.std():.3f})")

# Membuat model SVM dan mengevaluasi pada data validasi
print("\nEvaluasi Model SVM pada Validation Dataset:")
print("=" * 40)
svm = SVC()  # Membuat model SVM
svm.fit(X_train, Y_train)  # Melatih model menggunakan data latih
predictions = svm.predict(X_validation)  # Membuat prediksi pada data validasi

# Menampilkan hasil evaluasi
print(f"Accuracy Score    : {accuracy_score(Y_validation, predictions):.3f}")  # Akurasi
print("Confusion Matrix :")
print(confusion_matrix(Y_validation, predictions))  # Matriks kebingungan

# Laporan klasifikasi (precision, recall, f1-score untuk setiap kelas)
print("\nClassification Report:")
print(classification_report(Y_validation, predictions, zero_division=1))

# Menampilkan informasi dataset
print("\nInformasi Dataset:")
print("=" * 40)
print(f"Dataset Shape: {dataset.shape}")  # Ukuran dataset (jumlah baris dan kolom)
print(f"Contoh Data:\n{dataset.head(5)}")  # Contoh 5 data pertama
print("\nDistribusi Kelas:")
print(dataset.groupby('Class').size())  # Jumlah data di setiap kelas
```
